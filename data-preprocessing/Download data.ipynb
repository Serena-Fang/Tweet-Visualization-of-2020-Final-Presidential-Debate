```python
# Retrieve Oct.21 tweetID
for j in range(21,24):
    day = str(j)
    head = 'https://raw.githubusercontent.com/echen102/us-pres-elections-2020/master/2020-10/us-presidential-tweet-id-2020-10-'
    end = '.txt'
    for i in range(24):
        if i < 10:
            num = '0'+ str(i)
        else:
            num = str(i)
        url = head + day + '-' + num + end
        res = requests.get(url, allow_redirects=True)
        s_head = 'clean_data/10-'
        s_end = '.csv'
        s_url = s_head + day + '-' + num + end
        with open(s_url,'wb') as file:
            file.write(res.content)
```

```python
id_21 = []
id_22 = []
id_23 = []
id_full = []
```

```python
for i in range(24):
    if i < 10:
        num = '0'+ str(i)
    else:
        num = str(i)
    url = 'clean_data/10-21-' + num + '.txt'
    file = open(url, "r")
    for line in file:
        ids = line.strip()
        id_21.append(ids)
        id_full.append(ids)
```

```python
for i in range(24):
    if i < 10:
        num = '0'+ str(i)
    else:
        num = str(i)
    url = 'clean_data/10-22-' + num + '.txt'
    file = open(url, "r")
    for line in file:
        ids = line.strip()
        id_22.append(ids)
        id_full.append(ids)
```

```python
for i in range(24):
    if i < 10:
        num = '0'+ str(i)
    else:
        num = str(i)
    url = 'clean_data/10-23-' + num + '.txt'
    file = open(url, "r")
    for line in file:
        ids = line.strip()
        id_23.append(ids)
        id_full.append(ids)
```
